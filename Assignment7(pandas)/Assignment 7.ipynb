Task18:Make new column name Total_cases that have a sum of Deaths,

```python
data["Total_Cases_Calculated"] = data["Deaths"] + data["Recovered"] + data["Active"]
```

```python
data.head()
```

Task19:Calculate Death_Rate for Each Country (Deaths per Total
Cases) by suing formula (Desths/Total_cases)*100 and save it in
Dath_Rate column
-

```python
data["Death_Rate"] = (data["Deaths"] / data["Total_Cases_Calculated"]) * 100
```

```python
data.head()
```

Task20:Identify countries where total cases are increasing but death

```python
low_death_rate_countries= data[(data["Death_Rate"] < 5) & (data["Total_Cases_Calculated"].diff() > 0)]["Country/Region"]
```
